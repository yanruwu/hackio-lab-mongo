```python
import pymongo
import pandas as pd
```

```python
# lo primero que tenemos que hacer es conectarnos con Mongo
cliente = pymongo.MongoClient()
print(cliente)

# como la conexión se ha realizado con éxito vamos a ver las bases de datos que tenemos
cliente.list_database_names()
```

```python
# companies = pd.read_json("datos/companies.json", lines = True)
```

```python
companies = cliente["EcommerceDB"]["Companies"]
```

### 1. Nombre 'Babelgum'. Recuperar solo campo ``name``

```python
df_1 = pd.DataFrame(companies.find({'name':'Babelgum'}, {'name':1, "_id":0}))
df_1
```

### 2. Empresas con más de 5000 empleados. Limit 20 y order por número de empleados.

```python
df_2 = pd.DataFrame(companies.find({'number_of_employees' : {'$gt' : 5000}}, {"name" : 1, "number_of_employees" : 1}).sort({'number_of_employees' : -1}).limit(20))
df_2
```

### 3. Fundadas entre 2000 y 2005, inclusive. Recuperar solo `name` y `founded_year`

```python
df_3 = pd.DataFrame(companies.find({'founded_year':{'$in' : [y for y in range(2000,2006)]}}, {"name": 1, "founded_year": 1, "_id":0}))
df_3.head()
```

### 4. Valoración > 100000000, fundadas antes de 2005. Recuperar solo campos ``name`` e ``ipo``

```python
val_cond = {'ipo.valuation_amount':{"$gt" : 100000000}}
found_cond = {'founded_year':{"$lt": 2005}}
df_4 = pd.DataFrame(companies.find({"$and": [val_cond, found_cond]}, {"name":1, "ipo":1, "_id":0}))
df_4.head(10)
```

### 5. Menos de 1000 empleados, fundadas antes de 2005. Ordenar por número de empleados, límite de 10.

```python
found_cond = {'founded_year':{'$lt': 2005}}
employ_cond = {'number_of_employees':{'$lt':1000}}

df_5 = pd.DataFrame(companies.find({'$and': [found_cond, employ_cond]}).sort({'number_of_employees':-1}).limit(10))
df_5
```

### 6. Todas las empresas que no incluyen el campo ``partners``

```python
df_6 = pd.DataFrame(companies.find({'partners':{'$exists':False}})) # Sin campo partners
df_6.head()
df_6_2 = pd.DataFrame(companies.find({'partners':{'$size':0}})) # Valores vacíos en partners
df_6_2.head()
```

### 7. todas las empresas que tienen un valor nulo en el campo `category_code`

```python
df_7 = pd.DataFrame(companies.find({'category_code':{'$size':0}}))
df_7
```

### 8. Empresas con al menos 100 empleados pero menos de 1000

```python
gt100 = {'number_of_employees':{'$gt': 100}}
lt1000 = {'number_of_employees':{'$lt' : 1000}}
df_8 = pd.DataFrame(companies.find({'$and':[gt100, lt1000]}))
df_8.head()
```

### 9. Ordenar todas las empresas por su precio de IPO en orden descendente.

```python
df_9 = pd.DataFrame(companies.find().sort({'ipo.valuation_amount':-1}))
df_9.head()
```

### 10. Recupera las 10 empresas con más empleados, ordenadas por el número de empleados.

```python
df_10 = pd.DataFrame(companies.find({},{'name':1, 'number_of_employees':1}).sort({'number_of_employees':-1}).limit(10))
df_10
```

### 11. Todas las empresas fundadas en el segundo semestre del año. Búsqueda a 1000 empresas.

```python
df_11 = pd.DataFrame(companies.find({'founded_month': {'$gt':6}},{'name':1, 'founded_month':1}).limit(1000))
df_11.head(10)
```

### 12. Antes del 2000, monto de adquisición superior a 10000000.

```python
founded_cond = {'founded_year':{'$lt':2000}}
acqu_cond = {'acquisition.price_amount': {'$gt':10000000}}
df_12 = pd.DataFrame(companies.find({'$and':[founded_cond, acqu_cond]}, {'name':1, 'acquisition.price_amount':1}))
df_12.head()
```

### 13. Todas las empresas adquiridas tras 2010, ordenado por monto de adquisición, recupera solo name y acquisition.

```python
df_13 = pd.DataFrame(companies.find({'acquisition.acquired_year': {'$gt':2010}}, {'name':1, 'acquisition':1, '_id':0}).sort({'acquisition.price_amount':-1}))
df_13.head()
```

### 14. Ordenar las empresas por founded year, recuperando name y founded year.

```python
df_14 = pd.DataFrame(companies.find({},{'name':1, 'founded_year':1, '_id':0}).sort({'founded_year':-1}))
df_14.head(10)
```

### 15. Todas las empresas fundadas en los primeros siete días del mes, incluyendo el séptimo. Ordenar por acquisition price en orden descendente. Limitar a 10.

```python
df_15 = pd.DataFrame(companies.find({'founded_day':{'$in':[d for d in range(1,8)]}},{'name':1, 'founded_day':1, 'acquisition.price_amount':1}).sort({'acquisition.price_amount':-1}).limit(10))
df_15
```

### 16. Todas las empresas en la categoría web con más de 4000 empleados. Ordenar por cantidad de empleados, asc.

```python
web_cond = {'category_code':'web'}
employ_cond = {'number_of_employees': {'$gt': 4000}}

df_16 = pd.DataFrame(companies.find({'$and': [web_cond, employ_cond]}, {'name':1, 'number_of_employees':1}).sort({'number_of_employees':1}))
df_16.head()
```

### 17. Todas las empresas cuyo monto de adquisición es superior a 10000000, moneda EUR.

```python
acqu_cond = {'acquisition.price_amount':{'$gt': 10000000}}
coin_cond = {'acquisition.price_currency_code':'EUR'}

df_17 = pd.DataFrame(companies.find({'$and':[acqu_cond, coin_cond]}, {'name':1, 'acquisition.price_amount':1, 'acquisition.price_currency_code':1}))
df_17
```

### 18. Todas las empresas adquiridas en el primer trimestre del año. Limita a 10, recupera solo name y acquisition.

```python
df_18 = pd.DataFrame(companies.find({'acquisition.acquired_month': {'$in': [m for m in range(1,4)]}}, {'name':1, 'acquisition':1}).limit(10))
df_18
```
